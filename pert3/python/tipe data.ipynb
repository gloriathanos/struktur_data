```python
# INT (integer)

num = 10
num2 = 10

print(num + num2)

# STRING

name = "hello"
print(name)
print(type(name))

# FLOAT

c = 10.5
print(c)
print(type(c))

# COMPLEX

d = 1j
print(d)
print(type(d))

# LIST

e = ["a","b","c"]
print(e)
print(type(e))

# TUPLE

f = ("a","b","c")
print(f)
print(type(f))

# RANGE

g = range(5)
print(g)
print(type(g))

# SET

h = {"ani", "budi"}
print(h)
print(type(h))

# DICT

i = {"nama" : "ani"}
print(i)
print(type(i))

# FROZENSET

j = frozenset({"ani", "budi"})
print(j)
print(type(j))

# BOOL

k = True
print(k)
print(type(k))

# BYTES

l = b"hello"
print(l)
print(type(l))

# BYTE ARRAY

m = bytearray(5)
print(m)
print(type(m))

# MEMORY VIEW

n = memoryview(bytes(5))
print(n)
print(type(n))

# NONE

o = None
print(o)
print(type(o))
```
